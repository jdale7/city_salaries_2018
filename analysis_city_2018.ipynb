↓ figuring out which city departments are earning the most cumulative over time ("premium_pay"). the following calculations do the same thing with regular_pay, other_pay and total_pay.

```python
citysal.groupby("dept").premium_pay.sum().reset_index().sort_values("premium_pay", ascending=False).astype(str)
```

```python
citysal.groupby("dept").regular_pay.sum().reset_index().sort_values("regular_pay", ascending=False).astype(str)
```

```python
citysal.groupby("dept").other_pay.sum().reset_index().sort_values("other_pay", ascending=False).astype(str)
```

```python
citysal.groupby("dept").total_pay.sum().reset_index().sort_values("total_pay", ascending=False).astype(str)
```
