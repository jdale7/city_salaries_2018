↓ Figuring out that we have to get rid of the '$' figure and commas so we can do math on the numeric columns:

```python
citysal['regular_pay'] = citysal['regular_pay'].str.replace('\$', '')
```

```python
citysal['regular_pay'] = citysal['regular_pay'].str.replace('\,', '').astype(float)
```

```python
citysal.head()
```

```python
citysal.dtypes
```

Doing the rest:

```python
citysal['premium_pay'] = citysal['premium_pay'].str.replace('\$', '')
citysal['other_pay'] = citysal['other_pay'].str.replace('\$', '')
citysal['total_pay'] = citysal['total_pay'].str.replace('\$', '')
```

```python
citysal['premium_pay'] = citysal['premium_pay'].str.replace('\,', '').astype(float)
citysal['other_pay'] = citysal['other_pay'].str.replace('\,', '').astype(float)
citysal['total_pay'] = citysal['total_pay'].str.replace('\,', '').astype(float)
```

Looking at the city's 2018 budget, we know it had budgeted $219,480,596 for employee salaries and wages (not including benefits like healthcare). 

```python
city_salary_budget = 219480596
```

So, was the city over or under budget in terms of employee pay?

```python
(city_salary_budget - (citysal.total_pay.sum())).round(2)
```

It looks like the city was over budget by about $16.2 million in terms of employee pay.

We know from years past that employees of the city's Public Safety department, made up of the bureaus of police, fire, emergency medical services and animal care and control, earn a lot of overtime. Let's take a look at how much overtime those employees earned in 2018.

```python
police = citysal[citysal.dept == 'PS - Police Bureau']
```

```python
fire_dept = citysal[citysal.dept == 'PS - Fire Bureau']
EMS = citysal[citysal.dept == 'PS - Emergency Medical Services']
```

```python
police.premium_pay.sum()
```

```python
fire_dept.premium_pay.sum()
```

```python
EMS.premium_pay.sum().round(2)
```

Now let's look at the rest of the city departments. We can see in the table that the police, fire and EMS bureaus earned the most overtime pay. The calculations that follow look at each city department in terms of total regular_pay, other_pay and total_pay.

```python
citysal.groupby("dept").premium_pay.sum().round().reset_index().sort_values("premium_pay", ascending=False).astype(str)
```

```python
citysal.groupby("dept").regular_pay.sum().round().reset_index().sort_values("regular_pay", ascending=False).astype(str)
```

```python
citysal.groupby("dept").other_pay.sum().round().reset_index().sort_values("other_pay", ascending=False).astype(str)
```

```python
citysal.groupby("dept").total_pay.sum().round().reset_index().sort_values("total_pay", ascending=False).astype(str)
```
